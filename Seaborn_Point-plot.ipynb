```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Input data
data = {
    "DF": [58.39, 75.43, 60.4],
    "UF": [95.3, 101, 84.43],
    "RF": [61.58, 70.78, 81.53]
}
df = pd.DataFrame(data, index=["AI1", "AI2", "AI3"])

print(df)
```

```python
# NI activity data
data_NI = {
    "DF": [51.02, 41.86, 34.43],
    "UF": [60.89, 48.92, 54.35],
    "RF": [95.1, 99.2, 80.1]
}
df_NI = pd.DataFrame(data_NI, index=["NI1", "NI2", "NI3"])

print(df_NI)
```

```python
# Reshape both AI and NI data into long format
df_AI_long = df.reset_index().melt(id_vars="index", var_name="Stage", value_name="Activity")
df_AI_long["Type"] = "AI activity"

df_NI_long = df_NI.reset_index().melt(id_vars="index", var_name="Stage", value_name="Activity")
df_NI_long["Type"] = "NI activity"

# Combine datasets
df_combined = pd.concat([df_AI_long, df_NI_long])
df_combined.rename(columns={"index": "Replicate"}, inplace=True)
```

```python
# Plot with seaborn using red and blue colors
plt.figure(figsize=(6,5))
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['pdf.fonttype'] = 42  # Important for editable text in Illustrator
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
sns.pointplot(
    data=df_combined, x="Stage", y="Activity", hue="Type",
    errorbar="se", capsize=0.1, markers=["o","D"], 
    linestyles=["-","-"], palette=["#b71811","#118ab2"], linewidth = 2, markersize = 8
)

plt.ylabel("Enzyme acivities\n(U hexose/min/g FW)", fontsize = 18)
plt.xlabel("")
plt.title("")
plt.legend(title="")
ax = plt.gca()  # get current axis
ax.spines["bottom"].set_color("#5d637f")
ax.spines["left"].set_color("#5d637f")
ax.spines["bottom"].set_linewidth(2)
ax.spines["left"].set_linewidth(2)
plt.tight_layout()
plt.tick_params(axis='x', labelsize=18, color='#5d637f', length=6, width=2)
plt.tick_params(axis='y', labelsize=18, color='#5d637f', length=6, width=2)
# Save as PDF
plt.savefig("Enzyme_AI_NI_04.pdf", format="pdf", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['pdf.fonttype'] = 42  # Important for editable text in Illustrator
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
sns.lineplot(
    data=df_combined, x="Stage", y="Activity", hue="Type",
    errorbar="sd",  # standard error band
    marker="o", markersize=8, linewidth=2,
    palette=["#b71811","#118ab2"], linestyle="-"
)

plt.ylabel("Enzyme acivities\n(U hexose/min/g FW)", fontsize = 18)
plt.xlabel("")
plt.title("")
plt.legend(title="")
ax = plt.gca()  # get current axis
ax.spines["bottom"].set_color("#5d637f")
ax.spines["left"].set_color("#5d637f")
ax.spines["bottom"].set_linewidth(2)
ax.spines["left"].set_linewidth(2)
plt.tight_layout()
plt.tick_params(axis='x', labelsize=18, color='#5d637f', length=6, width=2)
plt.tick_params(axis='y', labelsize=18, color='#5d637f', length=6, width=2)
# Save as PDF
plt.savefig("Enzyme_AI_NI_03.pdf", format="pdf", bbox_inches='tight')
plt.show()
```
